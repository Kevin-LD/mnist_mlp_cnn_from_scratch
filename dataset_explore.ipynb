```python
import numpy as np
from struct import unpack
import gzip
import matplotlib.pyplot as plt
```

```python
path = r'./dataset/MNIST/train-images-idx3-ubyte.gz'

with gzip.open(path, 'rb') as f:
        magic, num, rows, cols = unpack('>4I', f.read(16))
        imgs=np.frombuffer(f.read(), dtype=np.uint8).reshape(num, 28*28)

print(magic)
print(num)
print(rows)
print(cols)
print(imgs.shape)
```

```python
print(imgs[0].max())
print(imgs[0].min())
```

```python
plt.figure(figsize=(10, 10), facecolor='white') 

for i in range(100):
    plt.subplot(10, 10, i + 1)
    plt.imshow(imgs[i].reshape(rows, cols), cmap='gray')
    plt.axis('off')

plt.subplots_adjust(left=0.02, right=0.98, bottom=0.02, top=0.98, wspace=0.05, hspace=0.05)

plt.show()
```
